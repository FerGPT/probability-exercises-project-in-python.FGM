## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO

print ("Planteamos dos estrategias diferentes:")

print ("\nN1")
# Lest design a simulation where 2 dices are thrown :)

import numpy as np

dado1 = np.random.randint(1, 7, 1000)
dado2 = np.random.randint(1, 7, 1000)

condiciones_cumplidas = 0
resultado = dado1 + dado2

for i in resultado:
    if (i % 2 == 0) or (i > 7):
        condiciones_cumplidas += 1
        
probabilidad = condiciones_cumplidas / 1000

print ("Calculated probability using a conditional loop:", probabilidad)

#--------------------------------------------------------------------------------------------------------
print ("\nN2:")


n_simulations = 5
# Generation of random integers from 1 to 6:
dice1 = np.random.randint(1, 7, size=n_simulations)
dice2 = np.random.randint(1, 7, size=n_simulations)

total = dice1 + dice2

success_greater = total > 7
success_even = (total % 2 == 0)
success_total = success_greater + success_even

probability = np.sum(success_total) / n_simulations

print("Calculated probability using vectors and utilising the property of sum of booleans as 'OR'. The probability of the getting a number equal or greater than 7 is approximately:", probability)
print ("\nThe folowing sum of booleans demonstrate that they behave like an 'OR':")
print (success_even)
print (success_greater)
print (success_total)
print (dice1)
print (dice2)


```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import numpy as np
import random

# Create the box of balls
ball_box = ["White"] * 10 + ["Red"] * 20 + ["Green"] * 30

# Run the experiment 1000 times
n_simulations = 1000
success_count_1 = 0  # For 3 white and 2 red
success_count_2 = 0  # For all same color

for _ in range(n_simulations):

    # Drawing 5 balls with replacement. Using "np.random.choice"
    # CHOICE : is optimized for speed and memory usage, making it suitable for large-scale simulations and data generation tasks.

    results = np.random.choice(ball_box, size=5, replace=True)

    # Count occurrences of each color
    white_count = np.sum(results == "White")
    red_count = np.sum(results == "Red")
    green_count = np.sum(results == "Green")

    # Condition 1: 3 white and 2 red
    if white_count == 3 and red_count == 2:
        success_count_1 += 1

    # Condition 2: All of the same color
    if white_count == 5 or red_count == 5 or green_count == 5:
        success_count_2 += 1

# Calculate probabilities
prob_1 = success_count_1 / n_simulations
prob_2 = success_count_2 / n_simulations

print("Probability of 3 white and 2 red:", prob_1)
print("Probability of all the same color:", prob_2)

# TODO
```
